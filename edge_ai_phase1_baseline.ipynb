```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(224, 224, 3)), # Example input shape
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(8, activation='softmax') # Assuming 8 classes based on previous cells
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.loss, model.optimizer
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
```

```python
DATASET_PATH = "/content/drive/MyDrive/EdgeAI_Dataset"

train_dir = f"{DATASET_PATH}/train"
val_dir   = f"{DATASET_PATH}/val"
test_dir  = f"{DATASET_PATH}/test"
```

```python
train_gen = ImageDataGenerator(
    preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input,
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

val_test_gen = ImageDataGenerator(
    preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input
)
```

```python
train_data = train_gen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="categorical",
    shuffle=True
)

val_data = val_test_gen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="categorical",
    shuffle=False
)

test_data = val_test_gen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode="categorical",
    shuffle=False
)
```

```python
print(train_data.class_indices)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=25,
    callbacks=[early_stop],
    verbose=1
)
```

```python
import numpy as np

test_data.reset()

y_prob = model.predict(test_data, verbose=1)
y_pred = np.argmax(y_prob, axis=1)
y_true = test_data.classes
class_names = list(test_data.class_indices.keys())
```

```python
from sklearn.metrics import accuracy_score

test_accuracy = accuracy_score(y_true, y_pred)
print("Overall Test Accuracy:", test_accuracy)
```

```python
from sklearn.metrics import classification_report
import pandas as pd

report = classification_report(
    y_true,
    y_pred,
    target_names=class_names,
    output_dict=True
)

metrics_df = pd.DataFrame(report).transpose()
metrics_df = metrics_df.loc[class_names, ["precision", "recall", "f1-score"]]

metrics_df
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=class_names,
    yticklabels=class_names
)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")

plt.tight_layout()
plt.savefig("confusion_matrix.png", dpi=300)
plt.show()
```

```python
# Load test dataset
import tensorflow as tf

test_ds = tf.keras.utils.image_dataset_from_directory(
    "/content/drive/MyDrive/EdgeAI_Dataset/test", # Corrected path
    image_size=(224, 224),
    batch_size=16,
    shuffle=False
)

class_names = test_ds.class_names
print("Classes:", class_names)
```

```python
# Run inference on test set
y_true = []
y_pred = []

for images, labels in test_ds:
    preds = model.predict(images)
    y_true.extend(labels.numpy())
    y_pred.extend(tf.argmax(preds, axis=1).numpy())
```

```python
# Single image inference simulation
import numpy as np

images, labels = next(iter(test_ds))
sample_img = images[0:1]
true_label = class_names[labels[0]]

pred = model.predict(sample_img)
pred_class = class_names[np.argmax(pred)]

print("Actual Class:", true_label)
print("Predicted Class:", pred_class)
print("Probabilities:", pred[0])
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt="d",
            xticklabels=class_names,
            yticklabels=class_names,
            cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
print(classification_report(
    y_true, y_pred,
    target_names=class_names
))
```
